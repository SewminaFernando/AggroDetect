## Setting Paths with Label from audio Dataset

### Import Libraries

```python
import os
import pandas as pd
```

```python
# Male dataset paths
male_aggressive_path = "Audio Dataset\\Male\\Aggressive"
male_non_aggressive_path = "Audio Dataset\\Male\\Non-Aggressive"
# Female dataset paths
female_aggressive_path = "Audio Dataset\\Female\\Aggressive"
female_non_aggressive_path = "Audio Dataset\\Female\\Non-Aggressive"
```

```python
# Save the path to audio and their labels, gender in a dataframe

female_agg_dir_list = os.listdir(female_aggressive_path)
path = []
aggressiveness = []
gender = []
for i in female_agg_dir_list:
    path.append(female_aggressive_path + '\\' + i)
    aggressiveness.append(1)
    gender.append(0)

female_agg_df = pd.DataFrame({'path': path, 'aggressiveness': aggressiveness, 'gender':gender})
female_agg_df.head()
```

```python
female_non_agg_dir_list = os.listdir(female_non_aggressive_path)
path = []
aggressiveness = []
gender = []
for i in female_non_agg_dir_list:
    path.append(female_non_aggressive_path + '\\' + i)
    aggressiveness.append(0)
    gender.append(0)

female_non_agg_df = pd.DataFrame({'path': path, 'aggressiveness': aggressiveness, 'gender':gender})
female_non_agg_df.head()
```

```python
male_agg_dir_list = os.listdir(male_aggressive_path)
path = []
aggressiveness = []
gender = []
for i in male_agg_dir_list:
    path.append(male_aggressive_path + '\\' + i)
    aggressiveness.append(1)
    gender.append(1)

male_agg_df = pd.DataFrame({'path': path, 'aggressiveness': aggressiveness, 'gender':gender})
male_agg_df.head()
```

```python
male_non_agg_dir_list = os.listdir(male_non_aggressive_path)
path = []
aggressiveness = []
gender = []
for i in male_non_agg_dir_list:
    path.append(male_non_aggressive_path + '\\' + i)
    aggressiveness.append(0)
    gender.append(1)

male_non_agg_df = pd.DataFrame({'path': path, 'aggressiveness': aggressiveness, 'gender':gender})
male_non_agg_df.head()
```

```python
# Concatenate all the dataframes
data = pd.concat([female_agg_df,female_non_agg_df,male_agg_df,male_non_agg_df], ignore_index=True, axis=0)

# Save the dataframe to a csv file
data.to_csv('Data\\Audio_Paths.csv', index=False)

data.head()
```
